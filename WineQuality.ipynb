```python
import pandas as pd
from google.colab import files, drive
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://drive.usercontent.google.com/u/0/uc?id=1TI-3xwTgO1hPDUbYta3pz6D3NolkJXv0&export=download', sep=';')
```

```python
df
```

```python
# Check empty values
df.isna().any().any()
```

```python
# Max wine quality
df.quality.max()
```

```python
df.describe()
```

```python
correlation = df.corr(method="pearson").round(1)

sns.heatmap(correlation, annot=True)
```

```python
# most significant columns: alcohol, sulphates, citric acid, fixed acidity
df2 = df[['alcohol', 'sulphates', 'citric acid', 'fixed acidity', 'quality']]
df2
```

```python

sns.pairplot(df2, hue='quality')
plt.show()
```

```python
# filter wines by quality rating of 7 and above and plot
df3 = df2[df2['quality'] >= 7]
sns.pairplot(df3, hue='quality')
plt.show()
```
